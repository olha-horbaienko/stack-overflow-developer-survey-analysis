```python
# TASK 3: WORK EXPERIENCE ANALYSIS (Завдання 3 Статистичний аналіз досвіду респондентів)
# Calculate mean, median and mode for respondents' work experience.
```

```python
# Перевіримо колонку WorkExp
df['WorkExp'].head()
```

```python
# Convert WorkExp to numeric (перетворення у числовий формат)
work_exp = pd.to_numeric(df['WorkExp'], errors='coerce')

# Check missing values (перевірка пропущених значень)
work_exp.isna().sum()

# Calculate mean (середнє значення)
mean_exp = work_exp.mean()

# Calculate median (медіана)
median_exp = work_exp.median()

# Calculate mode (мода)
mode_exp = work_exp.mode()[0]

pd.Series({
    'Mean': mean_exp,
    'Median': median_exp,
    'Mode': mode_exp
})
```

```python
# Result:
# The analysis shows the central tendency of respondents' work experience using mean, median, and mode.
# Аналіз показує центральну тенденцію досвіду роботи респондентів за допомогою середнього значення, медіани та моди.
```

```python
# TASK 5: PYTHON POPULARITY (Завдання 5 Визначення популярності Python)
# Calculate the percentage of respondents who use Python.
# Обчислити відсоток респондентів, які використовують Python.
```

```python
# Identify respondents who use Python (визначити респондентів, які використовують Python)
python_users = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)

# Count them (порахувати їх кількість)
python_count = python_users.sum()

# Total respondents (загальна кількість респондентів)
total = df['ResponseId'].nunique()

# Calculate percentage (обчислити відсоток)
python_percent = (python_count / total) * 100

# Format result as percentage (відформатувати як відсоток)
python_percent_rounded = round(python_percent, 2)

print(f"Python usage: {python_percent_rounded:.2f}%")
```

```python
# Result:
# The percentage of respondents who use Python is approximately 37.54%.
# Відсоток респондентів, які використовують Python, становить приблизно 37.39%.
```

```python
# TASK 7: GEOGRAPHIC SALARY ANALYSIS (Завдання 7 Географічний аналіз компенсації Python-розробників)
# Analyze the average and median compensation of Python developers by country.
# Проаналізувати середню та медіанну компенсацію Python-розробників за країнами.
```

```python
# Filter Python developers (відфільтрувати Python-розробників)
df_python = df[df['LanguageHaveWorkedWith'].str.contains('Python', na=False)]

# Remove missing salary values (прибрати пропущені значення зарплати)
df_python = df_python.dropna(subset=['ConvertedCompYearly'])

# Group by country and calculate statistics (групування по країнах)
result = df_python.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median'])

# Round values (округлення чисел)
result = result.round(2)

result
```

```python
# Result:
# The average and median compensation of Python developers were calculated for each country.
# Було обчислено середню та медіанну компенсацію Python-розробників для кожної країни.
```

```python
# TASK 8: EDUCATION ANALYSIS OF TOP EARNERS (Завдання 8 Аналіз освіти найбільш оплачуваних спеціалістів)
# Identify the education levels of the top 5 respondents with the highest compensation.
# Визначити рівні освіти 5 респондентів з найвищою компенсацією.
```

```python
# Remove missing salary values (прибрати пропущені значення)
df_clean = df.dropna(subset=['ConvertedCompYearly'])

# Sort by salary descending (сортування за зарплатою)
df_sorted = df_clean.sort_values(by='ConvertedCompYearly', ascending=False)

# Select top 5 respondents (вибрати топ-5)
top5 = df_sorted.head(5)

# Show their education levels (показати рівні освіти)

top5[['EdLevel', 'ConvertedCompYearly']]
```

```python
# Result:
# The education levels of the top 5 highest-paid respondents were identified.
# Some of the highest salary values may be outliers or unrealistic due to self-reported data.
# Було визначено рівні освіти 5 респондентів з найвищою компенсацією.
# Деякі з найвищих значень зарплати можуть бути викидами або нереалістичними через самостійне заповнення даних.
```

```python
# TASK 9: PYTHON POPULARITY BY AGE GROUP (Завдання 9 Аналіз популярності Python по віковим категоріям)
```

```python
# Identify respondents who use Python (визначаємо хто використовує Python)
df['UsesPython'] = df['LanguageHaveWorkedWith'].str.contains('Python', na=False)

# Group by age and calculate percentage (групування по віку та підрахунок %)
python_by_age = df.groupby('Age')['UsesPython'].mean() * 100

# Sort results (сортування результатів)
python_by_age.sort_values(ascending=False)

# Show results (вивід результату)
python_by_age = python_by_age.round(2)
python_by_age
```

```python
# Result:
# The percentage of respondents who use Python was calculated for each age group.
# Було розраховано відсоток респондентів, які використовують Python, для кожної вікової категорії.
```

```python
# TASK 10: INDUSTRY ANALYSIS OF HIGH-PAID REMOTE WORKERS (Завдання 10 Аналіз індустрій серед високооплачуваних віддалених працівників)
```

```python
# Calculate 75th percentile (75-й перцентиль)
q75 = df['ConvertedCompYearly'].quantile(0.75)

# Filter high-paid respondents (фільтр по зарплаті)
high_paid = df[df['ConvertedCompYearly'] > q75]

# Filter remote workers (фільтр віддалених)
remote_high_paid = high_paid[high_paid['RemoteWork'] == 'Remote']

# Count industries (підрахунок індустрій)
industry_counts = remote_high_paid['Industry'].value_counts()

industry_counts
```

```python
# Result:
# The most common industries among high-paid remote workers were identified.
# Було визначено найпоширеніші індустрії серед високооплачуваних віддалених працівників.
```
